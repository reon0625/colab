```python
import pandas as pd
df_feb=pd.read_csv('orders_feb.csv')
df_mar=pd.read_csv('orders_mar.csv')
df_jan=pd.read_csv('orders_jan.csv')
df_jan
```

```python
df_feb
```

```python
df_mar
```

```python
df=pd.concat([df_jan,df_feb,df_mar])
df
```

```python
df['sales_per_product']=df['price']*df['quantity']
df
```

```python
df_result=df.groupby('product')['sales_per_product'].sum().reset_index()
df_result
```
